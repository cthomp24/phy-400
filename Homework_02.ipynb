# **Question 1 Solution:**
#### Let $R_P$ equal the pump transition rate
#### Finding the rate equation $\dfrac{d N_3}{d t}$
$$\dfrac{d N_3}{d t} = -\gamma_{32} N_3(t) + R_P(N_0(t) - N_3(t))$$

#### Finding the rate equation $\dfrac{d N_2}{d t}$
$$\dfrac{d N_2}{d t} = \gamma_{32} N_3(t) + (-\gamma_{20} -\gamma_{21}) N_2(t)$$

#### Finding the rate equation $\dfrac{d N_1}{d t}$
$$\dfrac{d N_1}{d t} = \gamma_{21} N_2(t) - \gamma_{10} N_1(t)$$

#### Finding the rate equation $\dfrac{d N_0}{d t} = -\dfrac{d N_1}{d t} -\dfrac{d N_2}{d t} -\dfrac{d N_3}{d t}$
$$\dfrac{d N_0}{d t} = \gamma_{20} N_2(t) + \gamma_{10} N_1(t) -R_P(N_0(t) - N_3(t))$$

#### Finding the condition for population inversion
For population inversion to occur $N_2 > N_1$ to find the condition that satisfies this $\dfrac{N_2 - N_1}{N_3}>0$ needs to be found.
$$\dfrac{N_2 - N_1}{N_3} = \dfrac{\gamma_{32}}{\gamma_{20} +\gamma_{21}} - \dfrac{\gamma_{32}\gamma_{21}}{\gamma_{10}(\gamma_{20} +\gamma_{21})} > 0$$

$$\dfrac{\gamma_{21}}{\gamma_{10}} < 1$$

This means that in order for population inversion to occur $\gamma_{10} > \gamma_{21}$ needs to be true.

# **Question 3.1 Solution:**
Once the beam passes through the first mirror the electric field right before the second mirror is:
$$E_{cav} = E_{in} t_1 (-r_1 e^{-ikL})$$
After the beam reflects off the second missor the E field right before it hits the first mirror is:
$$E_{cav2} = E_{in} t_1 (-r_1 e^{-ikL})(-r_2 e^{-ikL}) = E_{in} t_1 (r_1 r_2 e^{-i2kL})$$
Every time the beam makes two full passes in the cavity the E field increases by $r_1 r_2 e^{-i2kL}$.\
This makes the equation for the E field for each number of full passes as follows (n is the number of round trips):
$$E_{cav} = E_{in} t_1(r_1 r_2 e^{-i2kL})^{n}$$
Using the geometric series you can then define the E field as:
$$E_{cav} = \dfrac{E_{in} t_1}{1 - r_1 r_2 e^{-i2kL}}$$

# **Question 4 Solution:**
$$\mathcal{F} = \dfrac{\mathcal{FSR}}{\nu_{\mathrm{FWHM}}}$$
$$\mathcal{F} = \dfrac{\dfrac{c}{2L}}{\dfrac{c}{\pi L} \arcsin({\dfrac{1-r_1r_2}{2\sqrt{r_1r_2}}})}$$
$$\mathcal{F} = \dfrac{\dfrac{\pi}{2}}{\arcsin({\dfrac{1-r_1r_2}{2\sqrt{r_1r_2}}})}$$
$$\mathcal{F} = \dfrac{\dfrac{\pi}{2}}{\dfrac{1-r_1r_2}{2\sqrt{r_1r_2}}}$$
$$\mathcal{F} = \dfrac{\pi \sqrt{r_1r_2}}{1-r_1r_2}$$
Using the binomial approximation ignoring higher terms and $r=\sqrt{1 - T - \mathcal{L}}$ you get:
$$r_1r_2 = 1-\dfrac{\mathcal{L}_{total}}{2}$$
Because we are assuming that we are in the high finesse limit we can assume that $\mathcal{L}_{total} << 1$ which causes:
$$\sqrt{r_1r_2} = 1$$
Plugging these into the equation for $\mathcal{F}$ we get:
$$\mathcal{F} = \dfrac{2 \pi}{\mathcal{L}_{total}}$$

# **Question 5 Solution**
$$\mathcal{F} = \dfrac{\pi \sqrt{r_1r_2}}{1-r_1r_2}$$
$$G_\mathcal{cav} = |g_\mathcal{cav}|^2 = |\dfrac{t_1}{1 - r_1 r_2}|^2 = \dfrac{t_1^2}{1 + r_1^2 r_2^2 - 2r_1 r_2}$$
#### (a)
For a critically coupled cavity $r_1 = r_2$\
This allows for $\mathcal{F}$ and $G_\mathcal{cav}$ to be simplified to:
$$\mathcal{F} = \dfrac{\pi r}{1-r^2}$$
$$G_\mathcal{cav} = \dfrac{t_1^2}{1 + r^4 - 2r^2} = \dfrac{T_1}{(r^2 -1)^2}$$
$$\dfrac{\mathcal{F}}{G_\mathcal{cav}} = \dfrac{\pi r(1-r^2)}{T_1}$$
Ignoring loss we can assume that $T_1 = 1-r^2$:
$$\dfrac{\mathcal{F}}{G_\mathcal{cav}} = \pi r$$
#### (b)
For a critically coupled cavity $r_2=1$\
This allows for $\mathcal{F}$ and $G_\mathcal{cav}$ to be simplified to:
$$\mathcal{F} = \dfrac{\pi \sqrt{r_1}}{1-r_1}$$
$$G_\mathcal{cav} = \dfrac{t_1^2}{1 + r_1^2 - 2r_1} = \dfrac{T_1}{(r_1-1)^2}$$
$$\dfrac{\mathcal{F}}{G_\mathcal{cav}} = \dfrac{\pi \sqrt{r_1} (r_1 - 1)}{T_1}$$
