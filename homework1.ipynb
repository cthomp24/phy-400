# **Question 2A Solution:**
Let\
$$f(x) = \sin(x)$$\
Then at x = 0\
$$f(0) = 1$$\
The first derivative $f'(x)$ is\
$$f'(x) = \cos(x)$$\
The first derivative evaluated at x = 0 is\
$$f'(0) = 0$$\
The second derivative $f''(x)$ is\
$$f''(x) = -\sin(x)$$\
The second derivative evaluated at x = 0 is\
$$f''(0) = -1$$\
The Taylor Expansion to the second order then becomes\
$$f(x) \approx f(0) + f'(0)x + \frac{1}{2!}f''(0)x^2$$
$$f(x) = 1 - \frac{1}{2!}x^2$$

```python
def binom(xx:float):
    return np.sin(xx)
```

```python
xx = np.linspace(-2*np.pi,2*np.pi)

taylor0 = 1
taylor1 = taylor0
taylor2 = taylor1 - 0.5*xx**2
```

```python
plt.figure(figsize=(10,6))

plt.plot(xx, binom(xx), label="Binomial Function for $f(x) = sin(x)$")
plt.plot(xx, taylor2, ls="--", label="Taylor 2")

plt.axvline(x=-1, label="$f''(x) = -1$")

plt.title("Binomial Approximations about $x = 0$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.grid()
plt.show()
```

# **Question 5A Solution:**
The real E field and the real B field for a spherical wave are as follows
$$E(r,t) = \dfrac{E_0}{r} e^{i(kr - \omega t)}\hat{n} = \dfrac{E_0}{r} \cos{(kr - \omega t)}\hat{n}$$
$$B(r,t) = \dfrac{1}{c r}E_0 e^{i(kr - \omega t)}(\hat{k} \times \hat{n}) = \dfrac{1}{c r}E_0 \cos{(kr - \omega t)}(\hat{k} \times \hat{n}) $$
Now solving for the Poynting vector $\boldsymbol{S}$
$$\boldsymbol{S} = \dfrac{1}{\mu_0}(\boldsymbol{E} \times \boldsymbol{B})$$
$$\boldsymbol{S} = \dfrac{1}{\mu_0}(\dfrac{E_0}{r} \cos{(kr - \omega t)}\hat{n} \times \dfrac{1}{c r}E_0 \cos{(kr - \omega t)}(\hat{k} \times \hat{n}))$$
$$\boldsymbol{S} = (\dfrac{E_0^{2}}{c \mu_0 r^{2}})(\cos{(kr - \omega t)})^{2}(\hat{k} \times \hat{n} \times \hat{n})$$
$$\boldsymbol{S} = (\dfrac{E_0^{2}}{c \mu_0 r^{2}})(\cos{(kr - \omega t)})^{2}(\hat{k})$$
